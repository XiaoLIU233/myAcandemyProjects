```python
# 'alpha' keyword to adjust the transparency level;
# 'size' argument is given in pixels
# 'cmap' means colormap and see:
# https://matplotlib.org/examples/color/colormaps_reference.html
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar();  # show color scale
```

```python
# errorbar
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k');
```

```python
plt.errorbar(x, y, yerr=dy, fmt='o', color='black',
             ecolor='lightgray', elinewidth=3, capsize=0);
```

```python
# Visualizing a Three-Dimensional Function
```

```python
# design a function z=f(x,y)
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
# Now let's look at this with a standard line-only contour plot:
plt.contour(X, Y, Z, colors='black');
```

```python
# more lines to be drawn20 equally spaced intervals within the data range:
plt.contour(X, Y, Z, 20, cmap='RdGy');
```

```python
plt.cm.
```

```python
# filled contour plot 
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar(); 
# creates an additional axis with labeled color information for the plot
```

```python
# plt.imshow
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image');
```
